Label encoding in machine learning is a preprocessing technique that converts categorical text data into numerical integers to make it machine-readable.

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Make']=le.fit_transform(df['Make'])
df['Make']
```

data wrangling
Data wrangling  is the iterative process of cleaning, structuring, and transforming raw, complex, or inconsistent data into a usable format for analysis, machine learning, or visualization

```python
from sklearn.preprocessing import LabelEncoder

le_model = LabelEncoder()
df['Model'] = le_model.fit_transform(df['Model'])

le_transmission = LabelEncoder()
df['Transmission'] = le_transmission.fit_transform(df['Transmission'])

le_fuel_type = LabelEncoder()
df['Fuel Type'] = le_fuel_type.fit_transform(df['Fuel Type'])

le_vehicle_class = LabelEncoder()
df['Vehicle Class'] = le_vehicle_class.fit_transform(df['Vehicle Class'])

x=df.drop('Vehicle Class', axis=1)
y=df['Vehicle Class']
```

MODELS

1.LINEAR REGRESSION
2.POLYNOMIAL REGRESSION
3.LASSO REGRESSION

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=32)
```

LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
train_predictions=model.predict(x_train)
test_predictions=model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
print('MAE for test data:',mean_absolute_error(y_test,test_predictions))
print('MAE for train data:',mean_absolute_error(y_train,train_predictions))
```

```python
from sklearn.metrics import r2_score
print('R2 for test data:',r2_score(y_test,test_predictions))
print('R2 for train data:',r2_score(y_train,train_predictions))
```

```python
from sklearn.metrics import mean_absolute_error
print('MSE for test data:',mean_absolute_error(y_test,test_predictions))
print('MSE for train data:',mean_absolute_error(y_train,train_predictions))
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
print('RMSE for test data:',np.sqrt(mean_squared_error(y_test,test_predictions)))
print('RMSE for train data:',np.sqrt(mean_squared_error(y_train,train_predictions)))
```

POLYNOMIAL REGRESSION

```python
from sklearn.metrics import r2_score
print('R2 for test data:',r2_score(y_test,test_predictions))
print('R2 for train data:',r2_score(y_train,train_predictions))
```
